## <a id="Variable" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Variables</a>
<span style="font-family: 'Great Vibes', cursive">
    
**Variable is a container used to store data in a program. It acts as a name assigned to a value, which can be changed or updated during execution.**



- Variables are created by assigning a value to a name using the `=` operator, such as x = 10.

- Python is **case-sensitive**, so age and Age are considered different variables.

- Variables can store different types of data, including integers, floats, strings, lists, tuples, and dictionaries.

- Python variables do not require explicit declaration; they are **dynamically typed**.

- The type of data stored in a variable can be checked using the `type()`function.

- Multiple variables can be assigned values simultaneously, such as **a, b, c = 1, 2, 3**.

- Using descriptive variable names improves code readability and maintainability.

- Following Python’s variable naming conventions, such as snake_case, makes code more standardized and readable.


```python
x = 10  # Storing an integer
print(x)
```

```python
y = 5.5  # Storing a float (decimal number)
print(y)
```

```python
name = "Alice"  # Storing a string
print(name)
```

## <a id="slicing" style="font-family: 'Great Vibes', cursive; font-size: 40px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">String Slicing</a>
<span style="font-family: 'Great Vibes', cursive">
    
Slicing in Python is a technique used to extract a portion of a sequence, such as a string, list, or tuple, by specifying a start and end index.

<span style="color: #0f52b4">**string[start:end:step]**</span>

- <span style="font-weight: bold; color: #0f52b4">start:</span> The index where slicing begins (default is 0).
- <span style="font-weight: bold; color: #0f52b4">stop:</span> The index where slicing stops (exclusive, meaning the element at this position is not included).
- <span style="font-weight: bold; color: #0f52b4">step:</span> The interval between indices (default is 1).




```python
quote = "it's a beautiful life"
print(quote[7:16:1]) 
```

```python
print(quote[1:5])  
```

```python
print(quote[7:])  
```

```python
print(quote[:2])  
```

```python
print(quote[7:16:2])  
```

## <a id="list" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; ">List</a>
<span style="font-family: 'Great Vibes', cursive">
A list is an ordered, mutable collection of elements that can store different data types, including integers, floats, strings, and even other lists. Lists are defined using square brackets [] and elements are separated by commas.   


Characteristics of a List:
- <span style="font-weight: bold; color: #0f52b4">Ordered:</span> The elements maintain their insertion order.

- <span style="font-weight: bold; color: #0f52b4">Mutable:</span> Elements can be changed, added, or removed.

- <span style="font-weight: bold; color: #0f52b4">Heterogeneous:</span> Can store different data types.

- <span style="font-weight: bold; color: #0f52b4">Indexable:</span>Supports indexing and slicing.

```python
my_list_1 = ['ahmad','nazanin','fateme']
print(my_list_1)
```

```python
my_list_2 = [1, 3.5, "Python", "hello"]
print(my_list_2)
```

## <a id="Indexing-and-Slicing-in-List" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Indexing and Slicing in List</a>
<span style="font-family: 'Great Vibes', cursive">
    
- In Python, lists are ordered collections that support indexing and slicing for accessing and modifying elements.

- Each item in a list has an index, starting from 0 for the first element.

- Negative indexing starts from the end, where -1 represents the last element.



```python
numbers = [10, 20, 30, 40, 50]
print(numbers[0])  
```

```python
print(numbers[-1])  
```

```python
# Slicing a list
fruits = ["apple", "banana", "cherry", "date"]
print(fruits[1:3])
```

```python
print(fruits[:2])   
```

```python
# Joining lists together
list1 = [1, 2, 3]
list2 = [4, 5, 6]
print(list1 + list2)
```

```python
# Modifying a list
colors = ["red", "green", "blue"]
colors[1] = "yellow"
print(colors) 
```

## <a id="Methods-of-Lists" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Methods of Lists</a>
<span style="font-family: 'Great Vibes', cursive">
    
- `len` (Lenth of a list)

- `append` (Adding an element)

- `insert` (Adding an element)

- `remove` (Removing an element)

- `pop` (Removing an element)

- `sort` (Sorting a list)

```python
fruits = ["apple", "banana", "cherry", "date"]
print(len(fruits))
```

```python
numbers = [1, 2, 3]
numbers.append(4)  # Adds an element to the end
print(numbers)  
```

```python
numbers.insert(1, 10)  # Inserts 10 at index 1
print(numbers)  
```

```python
animals = ["cat", "dog", "rabbit", "dog"]
animals.remove("dog")  # Removes the first occurrence of "dog"
print(animals)  
```

```python
popped = animals.pop(1) # Removes index 1
print(animals)
print(popped) 
```

```python
nums = [4, 1, 7, 3, 9]
nums.sort()  # Sorts in ascending order
print(nums)
```

```python
nums.sort(reverse= True) # Sorts in descending order
print(nums)
```
