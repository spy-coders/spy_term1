## <a id="if-Statement" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">if-Statement</a>
<span style="font-family: 'Great Vibes', cursive">
    
Executes code only if the condition is True:  
`if True:` --> `It runs ...`

```python
if 5 > 3:
    print('It is bigger !!!')
```

```python
hematocrit = 60
if hematocrit >= 54:
    print("You are dehydrated!!! ") 
```

```python
courses = ["Genetics", "Physiology", "Anatomy", "Immunolgy", "Pathology"]
if "Anatomy" in courses:
    print("Anatomy is an available course in this semester")
```

```python
differential_diagnosis = ['Ashtma', 'common cold', 'COPD', 'Sinusitis']
if "MI" not in differential_diagnosis:
    print('Your heart is working well')
```

## <a id="if-else Statement" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">if-else Statement</a>
<span style="font-family: 'Great Vibes', cursive">
    
Runs one block of code if the condition is True, and another if it's False.

```python
heart_rate = 123
if heart_rate > 100 :
    print('The patient has tachycardy')

else:
    print('The patient does not have tachy cardy')
```

```python
wbc_per_microliter = 6800
if wbc_per_microliter > 5000 and wbc_per_microliter < 10000:
    print("It's Normal!!! ")
else:
    print("You are facing a problem") 
```

## <a id="if-elif-else Statement" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">if-elif-else Statement</a>
<span style="font-family: 'Great Vibes', cursive">
    
Allows multiple conditions to be checked.

```python
heart_rate = 123
if heart_rate > 100 :
    print('The patient has tachycardy')
elif heart_rate < 60 :
    print('The patient has bradycardy')
else:
    print('The patient has normal heart rate')
```

#### **Introduction to Heart Failure (HF)**  
Heart failure (HF) occurs when the heart cannot pump enough blood to meet the body's needs. It can be broadly categorized into left-sided heart failure and right-sided heart failure, each with distinct symptoms and causes:
- **Left-Sided HF:** Lung-edema, Dilated LA(left atrium), Weakness, Shortness of Breath
  
- **Right-Sided HF:** elevated-JVP(jugular venous pressure), Swelling in abdomen and legs, Weakness, Hepatomegaly

Each symptom has one score:
- if score reaches three we confirm the diagnosis.
- if score reaches 2, we need extra examinations.
