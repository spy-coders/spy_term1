## <a id="Print" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">What is Loop?</a>
<span style="font-family: 'Great Vibes', cursive">

&#128013; In the early days of programming, performing repetitive tasks often meant writing the same lines of code multiple times—a tedious and error-prone process. This approach was not only inefficient but also made programs lengthy and difficult to maintain.  
To overcome this challenge, Python introduced loops, a fundamental programming construct that automates repetitive actions. Loops allow a task to be repeated dynamically based on specified conditions, making code more concise, readable, and easier to manage. By harnessing loops, programmers can efficiently handle large-scale repetitive operations, such as processing datasets, generating repetitive outputs, or iterating through collections, all while writing minimal code.


&#128013; Loops in Python are fundamental constructs used to automate repetitive tasks efficiently. They enable developers to iterate over sequences, process data dynamically, and control execution flow

&#128013; Python has two primitive loop command:   
1.  **While**  
2.  **For**

## <a id="While Loops" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">While Loops</a>
<span style="font-family: 'Great Vibes', cursive">

* A while loop in Python executes a block of code repeatedly **as long as a specified condition remains true.**
*  Unlike a for loop, which iterates over sequences like lists or ranges, a while loop continues until the condition changes.

```python
while condition:      
    # Code to execute while condition is True
```

```python
# Print -I am a student- 10 times
step = 1
while step <= 10:
    print('I am a student')
    step = step + 1
```

## <a id="Else in While Loops" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Else in While Loops</a>
<span style="font-family: 'Great Vibes', cursive">

* In Python, the `else` statement can be used with a `while` loop to execute a block of code only when the loop condition becomes false naturally.


```python
i = 1
while i <= 5:
    print(i)
    i = i + 1
else:
    print("Finished counting!")
```

```python
# See the difference - Thermometer
while True:
    tempreature = int(input())
    if tempreature > 37.5:
        print('the patient has fever')
    else:
        print('the patient does not have fever')
```

## <a id="For Loops" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">For Loops</a>
<span style="font-family: 'Great Vibes', cursive">

```python
# Basic Syntax
numbers = [11, 22, 33, 44, 55]
print(numbers[0])
print(numbers[1])
print(numbers[2])
print(numbers[3])
print(numbers[4])

# for num in numbers:
#   print(num)
```

```python
for num in numbers:
  print(num + 3)
```

```python
# Print 0-9
for num in range(0, 10): #do not contain 10
  print(num)
```

## <a id="Print" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Extra options on loops</a>
<span style="font-family: 'Great Vibes', cursive">

* `pass`does nothing—it’s just a placeholder when a loop or function must be syntactically complete but doesn’t need execution yet.
* `break` completely stops the loop and exits immediately.
* `continue`skips the current iteration and moves to the next.



```python
for i in [1,2,3,4,5]:
    pass
```

```python
for i in [10,15,20,21,35]:
    if i % 5 != 0:
        print(i,'is not dividabe')
        break
    print(i)
```

```python
for i in [10,15,20,21,35]:
    if i % 5 != 0:
        print(i,'is not dividabe')
        continue
    print(i)
```

```python
# Expire-time
heart_rate = [(89, '18:20'), (112, '18:40'), (84, '19:00'), (0, '19:20'), (0, '19:40'), (0, '20:00')]
for hr, time in heart_rate:
    if hr == 0:
        print('Expire time is : ', time)
        break
```

```python
BPs = [98, None, 110, 99, None, 100, 130]
abnormal = 0
for bp in  BPs:
    if bp is None:
        continue
    if bp > 120:
        abnormal = abnormal + 1

print('Number of high blood pressure reaadings in 7 time monitoring:', abnormal)
```

## <a id="Usage of _" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Usage of _</a>
<span style="font-family: 'Great Vibes', cursive">

*If we don't want to use the variable, we can use **_**

```python
# Print ten times 
for _ in range(10):
    print('I am a student')
```

```python
books = ['Harrison', 'Guyton', 'Gray']

for book in books:
  print('Harison')
```

```python
# we can do this instead:
for _ in books:
  print('Harison')
```
