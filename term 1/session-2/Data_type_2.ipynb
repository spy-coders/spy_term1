## <a id="Dictionary" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Dictionary</a>
<span style="font-family: 'Great Vibes', cursive">
    
- A dictionary in Python is an unordered collection of key-value pairs.
-  It's one of the most powerful built-in data structures because it allows fast lookups and efficient data storage.
-  You can create a dictionary using {} or the dict() constructor
- You access dictionary values using keys




```python
my_dict = {"Book": "Gray's Anatomy", "year": 2024, "author": "Gray"}
print(my_dict) 
```

```python
print(my_dict["Book"]) 
```

```python
my_dict["year"] = 2023  # Updating value
my_dict["country"] = "USA"  # Adding new key-value pair
print(my_dict)
```

## <a id="Dictionary Methods" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Dictionary Methods</a>
<span style="font-family: 'Great Vibes', cursive">

- Removing Elements
  - `pop(key)` removes a key and returns its value.
  - del deletes a key-value pair.
  - `clear()` removes all elements.
- Accessing keys and values
  - `keys()` returns a list of keys
  - `values()` returns a list of values
  - `items()` returns key-value pairs

```python
return_value = my_dict.pop("year")  # Removes 'year'
print(my_dict)
print(return_value)
```

```python
del my_dict["author"]  # Deletes 'author'
print(my_dict)
```

```python
my_dict.clear()  # Empties dictionary
print(my_dict)
```

```python
my_dict_new = {"Book" : "Guyton", "year" : 2020 , "author": "John E. Hall"}
print(my_dict_new.keys()) 
```

```python
print(my_dict_new.values()) 
```

```python
print(my_dict_new.items())  
```


## <a id="Mutable-vs-Immutable" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Mutable vs Immutable</a>
<span style="font-family: 'Great Vibes', cursive">


<span style="font-family: 'Great Vibes', cursive; font-size: 24px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">
1. Mutable
</span>
<span style="font-family: 'Great Vibes', cursive">
    
- Mutable objects can be modified after they are created.
- Examples: lists, dictionaries, sets
- Modifications such as adding, removing, or changing elements are allowed. 

<span style="font-family: 'Great Vibes', cursive; font-size: 24px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">
2. Immutable
</span>
<span style="font-family: 'Great Vibes', cursive">
    
- Immutable objects cannot be modified after creation.
- Examples: integers, floats, strings, tuples
- Any operation that tries to modify them creates a new object instead.

```python
numbers = [1, 2, 3]
numbers[0] = 10  # Modifying first element
print(numbers)   
```

```python
text = "Hello"
text[0] = "Y" 
```

```python
example = "I am a Medical Student!!!"
print(example.lower())
print(example)

medical_Books = ["Guyton", "Ganong", "Bern", "Harper"]
medical_Books.append("Leninger") # Modifies the original list
print(medical_Books) 
```
