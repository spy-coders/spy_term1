## <a id="if-Statement" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Arithmetic Operators</a>
<span style="font-family: 'Great Vibes', cursive">
Arithmetic operators are symbols used in programming and mathematics to perform basic calculations.



```python
# Use + for addition:
print(99 + 247)
```

```python
# Use - for subtraction:
print(99 - 247)
```

```python
# Use * for multiplication
print(99 * 247)
```

```python
# Use / for division
print(99 / 247)
```

```python
# Use % to get the remainder of a division
print(247 % 99)
```

```python
# Use // to get the quotient of a division
print(247 // 99)
```

```python
# Use ** for exponentiation (power)
print(247 ** 99)
```

## <a id="Comparison Operators" style="font-family: 'Great Vibes', cursive; font-size: 36px; font-weight: bold; color: #0f52b4; letter-spacing: 1px;">Comparison Operators</a>
<span style="font-family: 'Great Vibes', cursive">

In Python, comparison operators are used to compare two values and return a Boolean (True or False) based on the comparison.

```python
print(99 > 247)
```

```python
print(99 < 247)
```

```python
print(99 <= 247)
```

```python
print(99 >= 247)
```

```python
print(99 == 247)
```

```python
print(247 == 247)
```

```python
print(99 = 99)
```

```python
print(99 != 247)
```

```python
0.1 + 0.1 + 0.1 == 0.3
```

<span style="font-family: 'Great Vibes', cursive">
    
> **Note:** 
   the classic floating-point precision problem!  
The reason 0.1 + 0.1 + 0.1 == 0.3 evaluates to False in Python is due to how computers represent decimal numbers.
Computers store floating-point numbers in binary, which means certain decimal fractions (like 0.1) cannot be represented exactly. Instead, they're stored as an approximation.  
When you add up 0.1 + 0.1 + 0.1, the result isn't exactly 0.3, but a very close value like 0.30000000000000004. Since Python's == operator checks for exact equality, the comparison fails.</span>
